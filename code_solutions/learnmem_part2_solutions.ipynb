### 2.1 The energy function and learning rule

Consider a Hopfield network with $N$ neurons, denoted $S_i=\pm1$ for unit $i$, with dynamics

$$S_i = \text{sgn}\Big(\sum_jw_{ij}S_j\Big)$$

The synaptic weights between units $i$ and $j$ is denoted $w_{ij}$ and is symmetric, such that $w_{ij}=w_{ji}$, and $w_{ii} = 0$ for all $i$. The energy function for the network is defined as

$$H = -\frac{1}{2}\sum_{i=1}^N\sum_{j=1}^Nw_{ij}S_iS_j$$.

(a) Show that the dynamical update rule defined above will never increase the energy (Hint: consider an arbitrary neuron $i$ which starts in a state $S_i$, and then follows the update rule to move to a new state $S_i'$ -- calculate $H'-H$. Assume that all other neurons remain the same, i.e., $S_j' = S_j$ for $j\neq i$).

$\color{blue}{Solution}$:

First of all, following a one step update of neuron $i$, it is possible that its activity remains the same, i.e., $S_i' = S_i$. In that case, $H' = H$ and so $H' - H = 0$.

The more interesting case is the one in which neuron $i$ changes activity, so $S_i' = -S_i$. We write the difference between energies as

$$H' - H = -\frac{1}{2}\sum_{i=1}^N\sum_{j=1}^Nw_{ij}S_i'S_j' + \frac{1}{2}\sum_{i=1}^N\sum_{j=1}^Nw_{ij}S_iS_j$$

Since most of the neurons remain the same, many terms in the sum will cancel, and we will only be left with terms that include $S_i$ and $S_i'$:

$$H' - H = -\frac{1}{2}\sum_{j\neq i}w_{ij}S_i'S_j + \frac{1}{2}\sum_{j\neq i}w_{ij}S_iS_j = -\frac{1}{2}S_i'\sum_{j\neq i}w_{ij}S_j + \frac{1}{2}S_i\sum_{j\neq i}w_{ij}S_j$$

Then, we can replace $S_i'$ with $-S_i$:

$$H' - H = \frac{1}{2}S_i\sum_{j\neq i}w_{ij}S_j + \frac{1}{2}S_i\sum_{j\neq i}w_{ij}S_j = S_i\sum_{j\neq i}w_{ij}S_j$$

Since $w_{ii}=0$ for all neurons, the final expression above is simply the starting activity value of neuron $i$ multiplied by the input to neuron $i$. But according to the dynamical update rule defined above, in order for neuron $i$ to change its activity, this input should be of the opposite sign to $S_i$ itself. Thus, as a whole, this term should always be negative, and we see that the energy $H'$ will be lower than $H$ in this case.

(b) A reasonable criterion for the stability of each memory pattern is that it should correspond to a low energy state of the dynamics. Consider a Hopfield network with a single memory pattern stored, denoted $\xi_i=\pm1$. We define a new energy function

$$H = -\frac{1}{2N}\Big(\sum_iS_i\xi_i\Big)^2,$$

which ensures that the energy is lowest when the network state perfectly overlaps with this memory pattern (convince yourself that this is true). Compare this expression with the energy function defined above. How should the weights be set to ensure that this is a low-energy state? What about for an arbitrary number of patterns?

$\color{blue}{Solution}$:
    
$$H = -\frac{1}{2N}\Big(\sum_iS_i\xi_i\Big)^2 = -\frac{1}{2N}\Big( \sum_iS_i\xi_i \Big)\Big( \sum_jS_j\xi_j \Big) = -\frac{1}{2N}\sum_i\sum_j\xi_i\xi_jS_iS_j = -\frac{1}{2}\sum_i\sum_j\Big(\frac{1}{N}\xi_i\xi_j\Big)S_iS_j$$

Comparing this result with the energy function above, we see that we should set the weights as

$$w_{ij} = \frac{1}{N}\xi_i\xi_j$$

For the case of multiple patterns, our energy function should feature minima for each pattern:

$$H = -\frac{1}{2N}\sum_{\mu=1}^p\Big(\sum_iS_i\xi_i^\mu\Big)^2$$

We then follow the same steps as above:

$$H = -\frac{1}{2N}\sum_{\mu=1}^p\Big( \sum_iS_i\xi_i^\mu \Big)\Big( \sum_jS_j\xi_j^\mu \Big) = -\frac{1}{2N}\sum_i\sum_j\sum_\mu\xi_i^\mu\xi_j^\mu S_iS_j = -\frac{1}{2}\sum_i\sum_j\Big(\frac{1}{N}\sum_\mu\xi_i\xi_j\Big)S_iS_j$$

We then arrive at Hopfield's original Hebbian learning rule

$$w_{ij} = \frac{1}{N}\sum_{\mu=1}^p\xi_i^\mu\xi_j^\mu$$

NOTE: to impose no self-connections, $w_{ii}$ must be set to zero after this Hebbian rule is applied.

```python
# INPUT PATTERNS AND NOISY VERSIONS
# JUST RUN THIS BLOCK -- YOU DON'T NEED TO ADD ANYTHING

N = 25
p = 4

# patterns to be stored (I)
I = np.zeros((N,p))
D = np.array([0,1,2,3,6,9,11,14,16,19,21,22,23])
J = np.array([0,1,2,3,4,8,13,15,18,20,21,22])
C = np.array([1,2,3,4,5,10,15,21,22,23,24])
M = np.array([0,4,5,6,8,9,10,12,14,15,19,20,24])
I[D,0] = 1.
I[J,1] = 1.
I[C,2] = 1.
I[M,3] = 1.
I = 2*I - 1

# noisy patterns to be used as initial states (I_start)
I_start = np.zeros((N,p))
noisy_D = np.array([0,1,2,3,6,9,11,14,15,16,19,21,22,23])
noisy_J = np.array([0,1,2,3,4,8,13,15,18,20,22])
noisy_C = np.array([1,2,3,4,5,9,10,12,15,18,21,22,23,24])
noisy_M = np.array([1,4,6,8,9,10,12,15,19,20,24])
I_start[noisy_D,0] = 1.
I_start[noisy_J,1] = 1.
I_start[noisy_C,2] = 1.
I_start[noisy_M,3] = 1.
I_start = 2*I_start - 1

def reshape_pattern(x):
    return np.reshape(x,(5,5))

letters = ['D','J','C','M']

f = plt.figure(figsize=(7,4))
for i in range(4):
    plt.subplot(2,4,i+1)
    plt.imshow(reshape_pattern(I[:,i]),cmap='Greys')
    plt.xticks(())
    plt.yticks(())
    plt.title(letters[i],fontsize=16)
    plt.subplot(2,4,i+5)
    plt.imshow(reshape_pattern(I_start[:,i]),cmap='Greys')
    plt.xticks(())
    plt.yticks(())
plt.subplot(2,4,1)    
plt.ylabel("Memory Patterns",fontsize=14)
plt.subplot(2,4,5)
plt.ylabel("Initial Conditions",fontsize=14)
plt.tight_layout()
```

```python
# STEP 1 -- DEFINE WEIGHT MATRIX
# don't forget to remove self-connections!

def learn_weights(inpt):
    W = (1./N)*np.dot(inpt,inpt.T)
    W -= np.diag(np.diag(W))
    return W

W = learn_weights(I)
plt.imshow(W,cmap='RdBu')
plt.colorbar()
```

```python
# STEP 2 - SIMULATE

# write a function that takes a weight matrix and initial condition and simulates dynamics
# this function should return the resulting memory state
def simulate_hopfield(W,init):
    max_iters = 50
    c = 0
    diff = 1
    S = init
    y = np.zeros((N,max_iters,))
    while(diff != 0. and c<max_iters):
        S_next = np.sign(np.dot(W,S))
        diff = np.sum(S_next-S)
        S = S_next.copy()
        c += 1
    return S

# choose one of the noisy patterns as initial state
# try out all four! (x=0 to x=3)
x=2
S0 = I_start[:,x]
S = simulate_hopfield(W,S0)
```

### 2.3 Estimating the memory capacity with signal-to-noise analysis

Consider a Hopfield network with $N$ units, storing a set of $p$ patterns using the Hebbian learning rule $w_{ij} = \frac{1}{N}\sum_{\mu=1}^p\xi_i^\mu\xi_j^\mu$. We'd like to know how many patterns we can store in this network such that each one is a stable fixed point of the dynamics. There are many ways to do this, but here's one simple criterion: if all of the memory patterns are stable, then when the network is sitting in a memory state, it should stay there. In other words, all of the neurons that are currently active should remain active, and all of the neurons that are currently inactive should remain inactive.

To make this a bit more quantitative, let's consider the input to an arbitrary unit $i$ when the network is sitting in one of the memory states, say for pattern $1$. This input is defined as: $h_i^1 = \sum_{j=1}^Nw_{ij}S_j$. If we are around a memory state $1$, then we want our unit $i$ to remain in the correct state $\xi_i^1$. Therefore, we need the following inequalities to hold:

$$ \begin{cases}
h_i^1 > 0 &\text{for } \xi_i^1 = +1,\\
h_i^1 < 0 &\text{for } \xi_i^1 = -1.
\end{cases} $$

If we ensure that the probability of this input flipping sign is low, then we can be reasonably sure that our memories are stable! Let's now use this simple idea to estimate the memory capacity of the standard Hopfield network.

a) We now define a quantity $C_i^1 := \xi_i^1h_i^1$, which is simply the input to unit $i$ at pattern $1$ multiplied by the desired pattern state of unit $i$ for pattern $1$. Before we do any calculations, let's try to understand what this quantity $C_i^1$ means. It is simple to see that using $C_i^1$ instead of $h_i^1$ reduces the two inequalities defined above to a single one: $C_i^1 > 0$ (convince yourself that this is true!). Based on this, we can define a probability of making an error as:

$$P_{\text{err}} = \Pr(C_i^1 < 0)$$

b) Simplify the expression for $C_i^1$ such that it only depends upon the input patterns (i.e., only terms with $\xi_i^\mu$ ; Hint: (i) use the definition of the Hebbian learning rule for $w_{ij}$, and (ii) use the fact that the network is in memory state $1$, so $S_i = \xi_i^1$).

$\color{blue}{Solution}$:

$$C_i^1 = \xi_i^1h_i^1 = \xi_i^1\sum_jw_{ij}S_j = \xi_i^1\sum_jw_{ij}\xi_j^1 = \frac{1}{N}\sum_j\sum_\mu\xi_i^1\xi_i^\mu\xi_j^\mu\xi_j^1$$

c) Simplify the resulting equation further by pulling out part of the sum -- all terms for $\mu=1$. You should now have two terms, with the first being a single scalar number (the signal), and the second term depending on the other patterns (the noise, often called crosstalk). If the patterns are chosen randomly as $\pm1$ with equal probability, what distribution will the second term resemble? (Hint: it is a sum of a finite number of binary terms, or trials) Can we approximate it by an even simpler distribution? (Hint: Gaussian)

$\color{blue}{Solution}$:

Step 1: Simplify the equation into two terms: signal and noise

$$\begin{align}
C_i^1 &= \frac{1}{N}\sum_j\sum_\mu\xi_i^1\xi_i^\mu\xi_j^\mu\xi_j^1\\
&= \frac{1}{N}\sum_j\xi_i^1\xi_i^1\xi_j^1\xi_j^1 + \frac{1}{N}\sum_j\sum_{\mu\neq1}\xi_i^1\xi_i^\mu\xi_j^\mu\xi_j^1\\
&= \frac{1}{N}\sum_j(1) + \frac{1}{N}\sum_j\sum_{\mu\neq1}\xi_i^1\xi_i^\mu\xi_j^\mu\xi_j^1\\
&= 1 + \frac{1}{N}\sum_j\sum_{\mu\neq1}\xi_i^1\xi_i^\mu\xi_j^\mu\xi_j^1
\end{align}$$

Step 2: Approximate the second term with a probability distribution

Let's focus on the second term -- the noise term. Each term in the sum is the product of four independent binary random variables which take on values of $\pm1$ with equal probability. This means that overall, it is equivalent to a single random variable taking on values of $\pm1$ with equal probability. So what we have is a sum of $Np$ binary random variables -- this is in fact VERY similar to a Binomial distribution (except that we're dealing with $\{-1,+1\}$ random variables instead of $\{0,1\}$, but we can easily adapt to that!).

The Binomial distribution is a discrete probability distribution with two parameters $n_B$ and $p_B$. A binomial random variable $X$ (generally written $X \sim \mathcal{B}(n_B,p_B)$) describes the number of successes in $n_B$ repeated trials, where the probability of success on each trial is $p_B$. The distribution describes the probability of obtaining each outcome $x$ for a given parameter setting. In the case here, we have $n_B=Np$ trials in the sum, and each one has a probability of $p_B=0.5$ for success. The fact that we're using $\pm1$ variables means that the sum is not only counting the number of successes, but also subtracting the number of failures. To correct for this, we simply need to subtract the mean and multiply the binomial distribution by two:

$$X_{\pm1} \sim \mathcal{B}_{\pm1}(n_B,p_B) = 2(\mathcal{B}(n_B,p_B) - n_Bp_B),$$

Then, to make things even simpler, you may know that a binomial distribution is often very well approximated by a Gaussian distribution with the same mean and variance:

$$\mathcal{B}(n_B,p_B) \approx \mathcal{N}(n_Bp_B,n_Bp_B(1-p_B))$$

So what we now have is:

$$\begin{align}
C_i^1 &= 1 + \frac{2}{N}(\mathcal{B}(Np,0.5) - Np/2)\\
&\approx 1 + \frac{2}{N}(\mathcal{N}(Np/2,Np/4) - Np/2) = 1 + \frac{2}{N}\mathcal{N}(0,Np/4) = 1 + \mathcal{N}(0,p/N)
\end{align}$$

So the final result is a signal term, equal to $1$, and a noise term, which is a zero-mean Gaussian with variance that scales with the ratio of patterns stored divided by network size.

d) If we define an arbitrary cutoff $P_{\text{err}} < 0.01$, how many memories can be stored in the network? (Hints: (i) you should get something that scales with the network size $N$; (ii) since we are dealing with a Gaussian distribution, consider using the error function).

$\color{blue}{Solution}$:

According to our criterion defined above, we see that if we can estimate the probability of $C_i^1<0$ as a function of the number of memory patterns in the network, then we can find the maximum number of stored patterns such that $P_{\text{err}} \approx 0.01$. Based on the solution to the previous question, we see that $C_i^1$ will be less than zero if the Gaussian distribution is less than $-1$ (see figure below). It turns out this is equivalent to:

$$P_{\text{err}} = \frac{1}{2}\Big[1 - \text{erf}\big(1/\sqrt{2\sigma^2}\big)\Big] = \frac{1}{2}\Big[1 - \text{erf}\big(\sqrt{N/2p}\big)\Big],$$

where $\text{erf}$ is the error function (see figure below). Now, we can simply plug in different values for $p$ and $N$ and compute the probability of error. It is useful to defined a new parameter $\alpha = p/N$, often called the loading parameter. For $P_{\text{err}} \approx 0.01$, we arrive at a capacity of $\alpha = p/N \approx 0.15$.

### 2.4 Test memory capacity of Hopfield network

The analytical result found in the previous question involved several assumptions and shortcuts. Let's test it out by simulating Hopfield networks storing different numbers of patterns and then testing how stable the memories are. In order to test the stability of a memory, one simple way is to initialize the network in the memory state and simulate the dynamics until convergence (or until a maximum number of iterations is reached). If the network is still at or close to the same memory state, then we can be reasonably sure that the memory is stable enough. In order to measure how close the network is to a memory state, we will use a criterion called overlap, defined as

$$m^\mu = \frac{1}{N}\sum_{i=1}^N S_i\xi_i^\mu$$

for pattern $\mu$. The overlap is bounded between $-1$ and $1$, with $1$ meaning that the network is sitting at the memory, and $0$ meaning that it is uncorrelated with the memory. If we simulate the network starting at each memory, and the average overlap after simulation is close to $1$, then we can assume that all of the patterns are reasonably stable.

You will simulate this test below in three steps. First, you will need to write a function called $\texttt{overlap}$ which takes a weight matrix, initial state, and memory pattern, and returns the resulting overlap of the memory state following simulation of the dynamics. Use the function that you wrote before for the pattern completion question to simulate the dynamics. Next, you will need to write a function that takes in all memory patterns, and simulates each of them, returning the average overlap ($\texttt{avg_overlap}$). You should call the function $\texttt{overlap}$ in this one. Finally, loop through different numbers of patterns and call $\texttt{avg_overlap}$ function for different numbers of trials. Plot the results.

```python
### Step 1 -- write a function that takes a weight matrix, initial condition and memory pattern and return
# the overlap of the network state with that pattern following simulation
# use your previous code!

def overlap(W,init,I):
    S = simulate_hopfield(W,init)
    return np.dot(S,I)/N

### Step 2 -- iterate over different numbers of patterns and get the overlap for all memories
# write a function that gets average overlap for all patterns given a weight matrix and a set of patterns

def avg_overlap(W,Imat):
    p = Imat.shape[1]
    m = np.zeros((p,))
    for i in range(p):
        m[i] = overlap(W,Imat[:,i],Imat[:,i])
    return np.mean(m)

def generate_random_patterns(N,p):
    I = np.random.binomial(1,0.5,(N,p))
    return 2*I - 1

# Step 3 -- iterate over different numbers

N = 200
ntrials = 20
p_vals = range(1,40)
overlap_mat = np.zeros((len(p_vals),ntrials))
for i in range(len(p_vals)):
    for j in range(ntrials):
        Imat = generate_random_patterns(N,p_vals[i])
        W = learn_weights(Imat)
        overlap_mat[i,j] = avg_overlap(W,Imat)
```

```python
plt.figure()
plt.plot(p_vals,np.mean(overlap_mat,axis=1))
plt.plot(p_vals,0.97*np.ones((len(p_vals))),'--')
plt.legend(('average performance','cutoff'))
plt.ylabel('Average performance (overlap)',fontsize=14)
plt.xlabel('Number of stored patterns',fontsize=14)
```

### 2.5 Incremental learning and forgetting (programming exercise)

Up to now, we have assumed that a network can be trained with all patterns at once, and the weight matrix is set before simulation. However, in biology, memories are often learned incrementally over time. This poses a problem, however, because it means that the weight matrix needs to grow unbounded over time.

In this exercise you will simulate an incremental learning paradigm, in which a Hopfield network is trained with patterns slowly over time, and the weights are bounded between different values. The weight matrix should begin at zero, and should be updated according to

$$w_{ij}^\text{new} = w_{ij}^\text{old} + \eta\xi_i^\mu\xi_j^\mu,$$

with $w_{ij} \in [-A,A]$. Implement this below for a network of size $200$ with $90$ patterns trained over time. Observe what happens when the network is tested at a different number of patterns. Play around with the value of $A$. How does this affect performance?

```python
N = 200

# STEP 1 -- implement update weights function
def update_weights(W,A,eta,I):
    Wup = np.outer(I,I)
    Wup -= np.diag(np.diag(Wup))
    Wnew = W + eta*Wup
    Wnew[Wnew>A] = A
    Wnew[Wnew<-A] = -A
    return Wnew

# STEP 2 -- simulate a series of patterns and print what happens at different times
W = np.zeros((N,N))
max_pats = 90
Imat = generate_random_patterns(N,max_pats)
eta = 1./N
A = 5.0*eta

plt.figure()
m_stored = [20,40,60,80]
for i in range(max_pats):
    W = update_weights(W,A,eta,Imat[:,i])
    if i in m_stored:
        m = np.zeros((max_pats))
        for j in range(max_pats):
            m[j] = overlap(W,Imat[:,j],Imat[:,j])
        plt.plot(m)
plt.legend(m_stored)
plt.ylabel('Average performance (overlap)',fontsize=14)
plt.xlabel('Pattern number (learned in order)',fontsize=14)
```
